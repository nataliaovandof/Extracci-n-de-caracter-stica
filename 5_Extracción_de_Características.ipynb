### Reemplazar valores nulos

```python
valores_nulos=Micro_Retailer.isnull().sum()
valores_nulos
```

```python
Micro_Retailer =Micro_Retailer.fillna(round(Micro_Retailer.mean(),1))
```

```python
valores_nulos=Micro_Retailer.isnull().sum()
valores_nulos
```

```python
Micro_Retailer =Micro_Retailer.fillna(method='bfill') 
```

```python
valores_nulos=Micro_Retailer.isnull().sum()
valores_nulos
```

### Rango Intercuartílico

```python
fig = plt.figure(figsize =(15, 8))
Micro_Retailer.plot(kind='box', vert=False)
plt.title("Distribución del DataFrame")
plt.show() #dibujamos el diagrama
```

```python
y=Micro_Retailer

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido\n", Limite_Superior_iqr)
print()
print("Limite inferior permitido\n", Limite_Inferior_iqr)
```

```python
outliers_iqr= Micro_Retailer[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data_clean= Micro_Retailer[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean
```

```python
df_string = data_clean.select_dtypes(include=['object'])
```

```python
df_string.columns
```

```python
df_float = data_clean.select_dtypes(include=['float'])
```

```python
df_float.columns
```

```python
df_float1=df_float.copy()
df_float1=df_float1.fillna(round(df_float1.mean(),1))
df_float1
```

```python
fig = plt.figure(figsize =(15, 8))
df_float1.plot(kind='box', vert=False)
plt.title('Nueva distribución de valores del DataFrame')
plt.show() #dibujamos el diagrama
```

## Extracción de características

```python
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(Micro_Retailer)
```

### 1. Customer Relationship Tools

```python
#Obtengo un análisis univariado de una variable categórica en específico
table1= freq_tbl(Micro_Retailer['214_customer_relationship_tools'])
table1
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
table1= table1[table1['frequency']>1]
table1
```

```python
table1_index= table1.set_index('214_customer_relationship_tools')
table1_index
```

```python
table1_index['frequency'].plot(kind = 'bar', width=1, figsize=(10,4))
plt.title('Herramientas de relación con los clientes')
plt.xlabel('Tipo de Herramienta')
plt.ylabel('Frecuencia')
```
